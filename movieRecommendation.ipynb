```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#read the data
movies = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/tmdb_5000_credits.csv')
```

```python
#removing missing data
movies.dropna(inplace = True)
```

DATA PREPROCESSING

```python
#preprocess genres and keywords
import ast
def convert(obj):
  res = []
  for i in ast.literal_eval(obj):
    res.append(i['name'])
  return res
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
#preprocess the cast
def first10(obj):
  count = 0
  res = []
  for i in ast.literal_eval(obj):
    res.append(i['name'])
    count += 1
    if count == 10:
      return res

  return res
```

```python
movies['cast'] = movies['cast'].apply(first10)
```

```python
movies['cast'][0]
```

```python
#preprocess the crew
def fetch_director(obj):
  res = []
  crew_positions = ['Director', 'Producer', 'Executive Producer']
  for i in ast.literal_eval(obj):
    if i['job'] in crew_positions:
      res.append(i['name'])
  return res
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#preprocess the overview
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#concatenate the columns into one column
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
new_df = movies[['movie_id', 'title','tags']]
```

```python
#converting a list to a string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
#convert the tags in the string into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

MODEL BUILDING


```python
#Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
print(stop)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
print(vectors)
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#stemming
def stem(text):
  Y = []
  for i in text.split():
    #after stemming append it to list y
    Y.append(ps.stem(i))
  return " ".join(Y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:26]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)

```

```python
recommend("Harry Potter and the Philosopher's Stone")
```
